```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import accelerate
import bitsandbytes

model_name = 'mistralai/Mistral-7B-v0.1'

def load_quantized_model_and_tokenizer(model_name: str):
    """
    Load a quantized model and tokenizer with BitsAndBytesConfig for efficient inference.
    """
    bnb_config = BitsAndBytesConfig(
        torch_dtype=torch.bfloat16,  # Positional argument
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16,
        device_map="auto"
    )
    
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=bnb_config
    )
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token  # Set pad token if necessary
    
    return model, tokenizer

model, tokenizer = load_quantized_model_and_tokenizer(model_name)

# Create the pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    max_length=256  # Include max_length parameter here
)

def generate_text_with_pipeline(prompt):
    """
    Generate text using the loaded pipeline based on the given prompt.
    """
    sequences = pipe(prompt, max_new_tokens=100, do_sample=True, temperature=0.7, top_k=50, top_p=0.95, num_return_sequences=1)
    return sequences[0]['generated_text']

def main():
    user_input = input("Please enter your prompt, or press Enter to use a test prompt: ")
    prompt = user_input if user_input else "As a data scientist, can you explain the concept of regularization in machine learning?"
    
    print(f"Using prompt: {prompt}")
    generated_text = generate_text_with_pipeline(prompt)
    print("Generated Text:\n", generated_text)

if __name__ == "__main__":
    main()

```
